# Задание 1
Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным **16**.  
Найти доверительный интервал для оценки математического ожидания с надежностью **0.95**, если выборочная средняя **M = 80**,  
а объем выборки **n = 256**.

$$[X - Z_{a/2} * \frac{\sigma}{\sqrt{n}}, X + Z_{a/2} * \frac{\sigma}{\sqrt{n}}]$$

```python
Z = 1.96

sigma = 16

X = 80

n = 256

print("Доверительный интервал: [", X - Z*sigma/np.sqrt(n), ",", X + Z*sigma/np.sqrt(n), "]")
```

# Задание 2
В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:  
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1  
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X  
при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
t = 2.262 # Коэфф.Стьюдента находим по таблице (n=9, a=0.025)

n = 10

X = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]

M = np.mean(X) # Находим среднее значение X

X_df = pd.Series(X) # Переводим тип список в тип серии

X_df.var(ddof=1) # Считаем несмещенную оценку дисперсии

sigma = np.sqrt(X_df.var(ddof=1)) # Находим среднее квадратичное отклонение

# Считаем границы доверительного интервала
i1 = M - t*sigma/np.sqrt(n)
i2 = M + t*sigma/np.sqrt(n)

print("Доверительный интервал: [%.2f" % i1, ", %.2f" % i2, "]")
```

# Задание 3
Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр **17 мм**.  
Используя односторонний критерий с **α=0,05**, проверить эту гипотезу, если в выборке из **n=100** шариков средний диаметр
оказался равным **17.5 мм**, а дисперсия известна и равна **4 кв. мм**.

$$Zн = \frac{X - \mu_0}{\sigma / \sqrt{n}}$$

Гипотеза Н0: диаметр = 17 мм, µ =µ0  
Гипотеза Н1: диаметр > 17 мм, µ >µ0  

Критерий:  Z–тест

```python
Z = 1.645 # Находим по таблице при α=0,05

n = 100

M = 17

X = 17.5

D = 4

sigma = np.sqrt(D)

Z_n = (X - M)/(sigma/np.sqrt(n))

print("Zн =", Z_n)
```

Вывод: **Zн > Z**, следовательно **гипотеза H0 не верна**.

# Задание 4
Продавец утверждает, что средний вес пачки печенья составляет 200 г.  
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:  
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.  
Известно, что их веса распределены нормально.  
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

Гипотеза Н0: вес = 200 г, µ =µ0  
Гипотеза Н1: вес <> 200 г, µ <>µ0  

Критерий:  t–тест

```python
t = 3.25 # Находим по таблице при α=0,005

M0 = 200

n = 10

X = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]

X_df = pd.Series(X)

M = np.mean(X)

X_df.var(ddof=1) # Считаем несмещенную оценку дисперсии

sigma = np.sqrt(X_df.var(ddof=1)) # Находим средне-квадратичное отклонение

t_n = (M - M0)/(sigma/np.sqrt(n))

print("tн =%.3f" % t_n)
```

Вывод: **- t < tн < t**, следовательно **гипотеза H0 верна**.
