# Задание 1
Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов:  
Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.  
Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.  
Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
Footballers = np.array([173, 175, 180, 178, 177, 185, 183, 182])

Hockey_players = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])

Weightlifters = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
```

```python
x1 = np.array([5, 5, 5, 5, 5, 5, 5, 5])
x2 = np.array([10, 10, 10, 10, 10, 10, 10, 10, 10])
x3 = np.array([15, 15, 15, 15, 15, 15, 15, 15, 15, 15, 15])

plt.scatter(x1, Footballers)
plt.scatter(x2, Hockey_players)
plt.scatter(x3, Weightlifters)

plt.ylim([150, 200])
plt.xlim([0, 20])

plt.show()
```

В дисперсионном анализе проверяется гипотеза H0 о равенстве средних групповых значений количественного показателя.  
Применительно к нашей задаче, **гипотеза H0** состоит в том, что **средний показатель роста не зависит от вида спортивной группы**, в которой состоит спортсмен.

Выберем порог **α = 5%**.

Применим функцию для расчета значений дисперсионного анализа:

```python
stats.f_oneway(Footballers, Hockey_players, Weightlifters)
```

Поскольку **pvalue** составляет **1%**, а пороговое значение α = 5%, то делаем вывод о несостоятельности гипотезы H0, т.е. такой фактор как принадлежность к спортивной группе оказывает статистически значимое влияние на средний показатель роста спортсмена.
